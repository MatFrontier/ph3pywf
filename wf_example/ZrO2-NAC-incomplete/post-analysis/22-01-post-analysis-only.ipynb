```python
from fireworks import LaunchPad
from ph3pywf.workflows.core import wf_ph3py_post_analysis

c = {"t_min": 10,
     "t_max": 1001,
     "t_step": 5,
     "mesh": (11,11,11),
#      "born_filename": "BORN-phonopy",
     "kappa_properties_to_ignore": ["group_velocity", 
                                    "heat_capacity"],
    }

tag = "2021-08-25-12-55-50-661809"

print("Creating workflow...")
workflow = wf_ph3py_post_analysis(tag=tag,
                                  db_file_local="/home/jovyan/atomate/config/db.json",
                                  c=c,
                                 )
print("Created workflow")

### Create the launchpad and add our workflow #######################
print("Sending to LaunchPad...")
launchpad = LaunchPad.from_file("/home/jovyan/atomate/config/my_launchpad.yaml")
launchpad.add_wf(workflow)
print("Sent to LaunchPad")
print(">>Done<<")
```
